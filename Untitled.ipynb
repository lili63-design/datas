```python
#数据采集
import requests
import json
import time
import random
import re
import pandas as pd
from urllib.parse import quote

class BaiduDataJobSpider:
    def __init__(self, keyword_list):
        # 批量支持多关键词
        self.keyword_list = keyword_list
        self.base_url = "https://yiqifu.baidu.com/g/aqc"
        self.headers = {
            "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/120.0.0.0 Safari/537.36",
            "X-Requested-With": "XMLHttpRequest",
            "Referer": "https://yiqifu.baidu.com/g/aqc/joblist",
            "Accept": "application/json, text/javascript, */*; q=0.01",
            "Accept-Language": "zh-CN,zh;q=0.9,en;q=0.8",
            "Sec-Fetch-Mode": "cors",
            "Sec-Fetch-Site": "same-origin"
        }
        self.province_dict = {}   
        self.all_job_data = []   
        self.duplicate_jobid = set() 

    def get_all_province(self):
        """拉取全国省份编码字典"""
        url = f"{self.base_url}/getDistrictAjax"
        try:
            resp = requests.get(url, headers=self.headers, timeout=12)
            resp.encoding = "utf-8"
            if resp.status_code == 200:
                res_json = resp.json()
                if res_json.get("status") == 0:
                    raw_prov = res_json["data"]["provinceCode"]
                    if "100000" in raw_prov:
                        del raw_prov["100000"]
                    self.province_dict = raw_prov
                    return True
        except Exception as err:
            print(f"获取省份接口失败：{str(err)}")
        return False

    def parse_salary_data(self, salary_text):
        """统一解析薪资文本，保留平均月薪"""
        if not salary_text:
            return None
        clean_sal = salary_text.replace("元/月", "").replace("元/天", "").replace("元/小时", "")
        match = re.match(r"(\d+)-(\d+)", clean_sal)
        if match:
            low = int(match.group(1))
            high = int(match.group(2))
            avg = round((low + high) / 2, 2)
            return avg
        return None

    def clean_html_text(self, raw_str):
        return re.sub(r"<.*?>", "", raw_str).strip() if raw_str else ""

    def crawl_single_keyword_province(self, keyword, province_code, province_name):
        page_num = 1
        max_page_limit = 5 
        encoded_key = quote(keyword)

        while page_num <= max_page_limit:
            ajax_url = (
                f"{self.base_url}/joblist/getDataAjax"
                f"?q={encoded_key}&page={page_num}&pagesize=20&district={province_code}"
            )
            try:
                resp = requests.get(ajax_url, headers=self.headers, timeout=10)
                resp.encoding = "utf-8"
                if resp.status_code != 200:
                    print(f" {keyword} | {province_name} 第{page_num}页访问异常，跳过")
                    break
                res_json = resp.json()
                if res_json.get("status") != 0:
                    print(f" {keyword} | {province_name} 接口返回异常状态码")
                    break
                page_data = res_json["data"]
                job_list = page_data.get("list", [])
                total_count = page_data.get("total", 0)
                page_total = page_data.get("pageCount", 1)

                if not job_list:
                    print(f"{keyword} | {province_name} 第{page_num}页无岗位，结束该省份")
                    break

                # 逐条解析岗位
                for job in job_list:
                    jid = job.get("jobId")
                    if jid in self.duplicate_jobid:
                        continue
                    self.duplicate_jobid.add(jid)

                    sal_avg = self.parse_salary_data(job.get("salary"))
                    job_item = {
                        "检索关键词": keyword,
                        "岗位名称": self.clean_html_text(job.get("jobName")),
                        "省份": province_name,
                        "城市": job.get("city", ""),
                        "薪资原始文本": job.get("salary", ""),
                        "平均月薪": sal_avg,
                        "学历门槛": job.get("edu", ""),
                        "工作年限要求": job.get("exp", ""),
                        "企业名称": job.get("company", "")
                    }
                    self.all_job_data.append(job_item)

                print(f"{keyword} | {province_name} 第{page_num}页，本页新增{len(job_list)}条岗位，累计总数据{len(self.all_job_data)}")
                time.sleep(random.uniform(1.8, 3.2))
                if page_num >= page_total:
                    break
                page_num += 1

            except Exception as err:
                print(f"{keyword} | {province_name} 第{page_num}页抓取失败：{str(err)}")
                time.sleep(random.uniform(3,5))
                page_num += 1
    def run_all_crawl(self):
        """主执行入口：遍历全部关键词+全部省份"""
        if not self.get_all_province():
            print("省份列表加载失败")
            return
        for kw in self.keyword_list:
            print(f"\n========== 开始抓取关键词：{kw} ==========")
            for code, name in self.province_dict.items():
                self.crawl_single_keyword_province(kw, code, name)
            print(f"关键词 {kw} 全部省市抓取完成")
        df = pd.DataFrame(self.all_job_data)
        df.to_csv("采集数据集.csv", encoding="utf-8-sig", index=False)
        print(f"\n共采集 {len(self.all_job_data)} 条不重复岗位")
if __name__ == "__main__":
    target_keys = ["数据", "数据分析师", "数据工程师"]
    spider = BaiduDataJobSpider(keyword_list=target_keys)
    spider.run_all_crawl()
```
