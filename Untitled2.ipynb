```python
#聚类分析
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.dpi"] = 120 
plt.rcParams["axes.grid"] = False
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
color_list = ["#8DA9C4", "#D4A29C", "#B4C4B0", "#E8CFAF", "#A49CB7", "#D0A8B2"]
cluster_color = color_list[:3]
df = pd.read_csv("清洗后数据集.csv", encoding="utf-8-sig")
feature_cols = ["平均月薪","城市梯队编码","学历编码","年限编码"]
X_raw = df[feature_cols].copy()
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_raw)
# 肘部图
wcss = []
k_range = range(1,11)
for k in k_range:
    km = KMeans(n_clusters=k, init="k-means++", random_state=0)
    km.fit(X_scaled)
    wcss.append(km.inertia_)
plt.figure(figsize=(8,4))
plt.plot(k_range, wcss, marker="o", linewidth=2, color=color_list[4])
plt.scatter(3, wcss[2], marker="X", s=120, c="#c85a5a", zorder=5)
plt.xlabel("聚类数量K")
plt.ylabel("簇内误差平方和 WCSS")
plt.title("肘部法则确定最优聚类数量")
plt.tight_layout()
plt.show()
best_k = 3
kmeans = KMeans(n_clusters=best_k, init="k-means++", random_state=0)
df["聚类标签"] = kmeans.fit_predict(X_scaled)
# PCA降维
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
# PCA聚类边界散点图
x_min, x_max = X_pca[:,0].min()-1, X_pca[:,0].max()+1
y_min, y_max = X_pca[:,1].min()-1, X_pca[:,1].max()+1
xx, yy = np.meshgrid(np.linspace(x_min,x_max,1000), np.linspace(y_min,y_max,1000))
Z = kmeans.predict(pca.inverse_transform(np.c_[xx.ravel(), yy.ravel()])).reshape(xx.shape)
plt.figure(figsize=(7,5))
plt.contourf(xx, yy, Z, alpha=0.12, colors=cluster_color)
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=df["聚类标签"], palette=cluster_color, s=18)
centers_pca = pca.transform(kmeans.cluster_centers_)
plt.scatter(centers_pca[:,0], centers_pca[:,1], s=130, c="#c85a5a", label="簇质心", zorder=10)
plt.title(f"K-Means岗位聚类分群边界图(K={best_k})")
plt.xlabel(f"主成分1")
plt.ylabel(f"主成分2")
plt.legend()
plt.tight_layout()
plt.show()
# 岗位大类热力图 
plt.figure(figsize=(8,4))
job_cross = pd.crosstab(df["聚类标签"], df["岗位大类"])
sns.heatmap(job_cross, annot=True, fmt="d", cmap="Blues", alpha=0.8)
plt.title("聚类簇与岗位大类分布热力图")
plt.tight_layout()
plt.show()
# 多子图综合对比大图
fig, axes = plt.subplots(2,3, figsize=(15,9))
group = df.copy()
sns.violinplot(x="聚类标签", y="平均月薪", data=group, palette=cluster_color, ax=axes[0,0], inner="quartile")
axes[0,0].set_title("薪资分布")
cnt = group["聚类标签"].value_counts().sort_index()
axes[0,1].bar(cnt.index, cnt.values, color=cluster_color)
axes[0,1].set_title("样本数量")
exp_avg = group.groupby("聚类标签")["年限编码"].mean()
axes[0,2].bar(exp_avg.index, exp_avg.values, color=cluster_color)
axes[0,2].set_title("平均工作年限")
edu_cross = pd.crosstab(group["聚类标签"], group["标准化学历"], normalize="index")
edu_cross.plot(kind="bar", stacked=True, ax=axes[1,0], color=color_list)
axes[1,0].set_title("学历占比")
axes[1,0].legend(fontsize=6)
city_cross = pd.crosstab(group["聚类标签"], group["城市梯队"], normalize="index")
city_cross.plot(kind="bar", stacked=True, ax=axes[1,1], color=color_list[1:])
axes[1,1].set_title("城市梯队占比")
axes[1,1].legend(fontsize=6)
axes[1,2].axis("off")
fig.suptitle("三类岗位多维度特征综合对比", fontsize=14)
fig.tight_layout()
plt.show()
center_raw = scaler.inverse_transform(kmeans.cluster_centers_)
center_df = pd.DataFrame(center_raw, columns=feature_cols)
print("=====聚类簇特征中心=====")
print(center_df.round(2))
df.to_csv("聚类结果数据集.csv", encoding="utf-8-sig", index=False)
print("导出完成：聚类结果数据集.csv")
```
