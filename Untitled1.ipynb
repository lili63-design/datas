```python
#数据预处理
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
df_raw = pd.read_csv("采集数据集.csv", encoding="utf-8-sig")
df = df_raw.copy()
def blank_convert_nan(text):
    if isinstance(text, str) and text.strip() == "":
        return np.nan
    return text
df = df.applymap(blank_convert_nan)
df = df.dropna(subset=["岗位名称", "平均月薪"], how="all")
df = df.dropna(subset=["工作年限要求"])
df = df.reset_index(drop=True)
df = df[(df["平均月薪"] >= 1000) & (df["平均月薪"] <= 60000)]
df = df.reset_index(drop=True)
def clean_job_title(title):
    if pd.isna(title):
        return title
    title = re.sub(r"\(.*?\)", "", title)
    title = re.sub(r"\s+", "", title)
    return title.strip()
df["_内部标准化岗位名称"] = df["岗位名称"].apply(clean_job_title)
## 学历统一标准化分类
def standard_edu(edu_text):
    if pd.isna(edu_text) or edu_text == "不限":
        return "不限学历"
    if "大专" in edu_text:
        return "大专"
    if "本科" in edu_text:
        return "本科"
    if "硕士" in edu_text or "研究生" in edu_text:
        return "硕士及以上"
    if "博士" in edu_text:
        return "硕士及以上"
    return "其他学历"
df["标准化学历"] = df["学历门槛"].apply(standard_edu)
## 工作年限统一标准化分类
def standard_exp(exp_text):
    if pd.isna(exp_text) or exp_text == "不限":
        return "无经验要求"
    if "应届生" in exp_text or "1年以内" in exp_text:
        return "应届生/1年内"
    if "1-3年" in exp_text:
        return "1-3年"
    if "3-5年" in exp_text:
        return "3-5年"
    if "5-10年" in exp_text:
        return "5-10年"
    if "10年以上" in exp_text:
        return "10年以上"
    return "其他年限"
df["标准化工作年限"] = df["工作年限要求"].apply(standard_exp)
def clean_city_name(city_text):
    if pd.isna(city_text):
        return ""
    city = re.sub(r"^(.*?(省|自治区))", "", city_text)
    if city.endswith("市"):
        city = city[:-1]
    return city.strip()
df["_内部清洗后城市名"] = df["城市"].apply(clean_city_name)
# 城市梯队分类 
first_tier = {"北京","上海","广州","深圳"}
new_first_tier = {"杭州","苏州","成都","重庆","武汉","南京","天津","西安","长沙","青岛","宁波","合肥","佛山","东莞","郑州"}
def city_tier_classify(cleaned_city):
    if not cleaned_city:
        return "未知城市"
    if cleaned_city in first_tier:
        return "一线城市"
    elif cleaned_city in new_first_tier:
        return "新一线城市"
    else:
        return "二三线及以下城市"
df["城市梯队"] = df["_内部清洗后城市名"].apply(city_tier_classify)
df = df[df["城市梯队"] != "未知城市"]
df = df.reset_index(drop=True)
print("====城市梯队分布验证====")
print(df["城市梯队"].value_counts())
print("-" * 60)

# 衍生分层特征
## 划分岗位大类
def classify_job_type(row):
    kw = row["检索关键词"]
    title = row["_内部标准化岗位名称"]
    if kw == "数据分析师" or "分析师" in title:
        return "业务分析岗"
    elif kw == "数据工程师" or "开发" in title or "数仓" in title:
        return "技术开发岗"
    else:
        if "运营" in title or "采集" in title:
            return "基础数据岗"
        elif "销售" in title or "商务" in title:
            return "数据商务岗"
        else:
            return "综合数据岗"
df["岗位大类"] = df.apply(classify_job_type, axis=1)
## 薪资分层
def salary_segment(sal):
    if sal < 5000:
        return "低薪(<5k)"
    elif sal < 12000:
        return "中薪(5k-12k)"
    elif sal < 20000:
        return "高薪(12k-20k)"
    else:
        return "顶尖高薪(>20k)"
df["薪资分层"] = df["平均月薪"].apply(salary_segment)
## 类别特征数值编码
edu_encoder = LabelEncoder()
exp_encoder = LabelEncoder()
city_encoder = LabelEncoder()
job_encoder = LabelEncoder()
df["学历编码"] = edu_encoder.fit_transform(df["标准化学历"])
df["年限编码"] = exp_encoder.fit_transform(df["标准化工作年限"])
df["城市梯队编码"] = city_encoder.fit_transform(df["城市梯队"])
df["岗位大类编码"] = job_encoder.fit_transform(df["岗位大类"])
output_columns = [
    "检索关键词","岗位名称","省份","城市","城市梯队","城市梯队编码",
    "薪资原始文本","平均月薪","薪资分层",
    "学历门槛","标准化学历","学历编码",
    "工作年限要求","标准化工作年限","年限编码",
    "企业名称","岗位大类","岗位大类编码"
]
df_clean_final = df[output_columns]
df_clean_final.to_csv("清洗后数据集.csv", encoding="utf-8-sig", index=False)
print(f"\n数据预处理全部完成，最终有效样本：{len(df_clean_final)} 条")
print("已输出文件：清洗后数据集.csv")
print("\n====清洗后数据基础统计====")
print("1. 各岗位大类数量分布：")
print(df_clean_final["岗位大类"].value_counts())
print("\n2. 薪资分层分布：")
print(df_clean_final["薪资分层"].value_counts())
print("\n3. 学历要求分布：")
print(df_clean_final["标准化学历"].value_counts())
print("\n4. 工作年限分布：")
print(df_clean_final["标准化工作年限"].value_counts())
```

```python
#可视化
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from sklearn.cluster import KMeans
import matplotlib.colors as mcolors
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.dpi"] = 120 
plt.rcParams["axes.grid"] = False
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
color_list = ["#8DA9C4", "#D4A29C", "#B4C4B0", "#E8CFAF", "#A49CB7", "#D0A8B2"]
df = pd.read_csv("清洗后数据集.csv", encoding="utf-8-sig")

# 柱状图 - 各岗位大类岗位数量
plt.figure(figsize=(11,5))
job_cnt = df["岗位大类"].value_counts()
bars = plt.bar(job_cnt.index, job_cnt.values, color=color_list[:len(job_cnt)], width=0.6)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 20, f"{int(height)}", ha="center", fontsize=9)
plt.title("各岗位大类招聘岗位数量分布", fontsize=13, pad=12)
plt.xlabel("岗位大类", fontsize=10)
plt.ylabel("岗位数量", fontsize=10)
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
# 饼图 - 薪资分层占比
plt.figure(figsize=(7,7))
sal_cnt = df["薪资分层"].value_counts()
wedges, texts, autotexts = plt.pie(
    sal_cnt.values, labels=sal_cnt.index, autopct="%1.1f%%",
    colors=color_list, startangle=90, textprops={"fontsize":9}
)
plt.title("全量岗位薪资分层占比", fontsize=13, pad=12)
plt.tight_layout()
plt.show()
# 直方图 - 平均薪资数值分布 
plt.figure(figsize=(10,5))
plt.hist(df["平均月薪"], bins=22, color=color_list[0], alpha=0.75, edgecolor="white")
plt.title("岗位平均薪资数值分布直方图", fontsize=13, pad=12)
plt.xlabel("平均月薪（元）")
plt.ylabel("岗位频数")
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
#箱线图 - 不同城市梯队薪资对比 
plt.figure(figsize=(10,5))
city_order = ["一线城市", "新一线城市", "二三线及以下城市"]
box_data = [df[df["城市梯队"]==city]["平均月薪"] for city in city_order]
box = plt.boxplot(box_data, labels=city_order, patch_artist=True)
# 给箱体上色
for patch, c in zip(box["boxes"], color_list[:3]):
    patch.set_facecolor(c)
    patch.set_alpha(0.7)
plt.title("三大城市梯队薪资分布箱线图", fontsize=13, pad=12)
plt.xlabel("城市梯队")
plt.ylabel("平均月薪（元）")
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
# 堆叠柱状图 - 岗位大类×学历分布 
plt.figure(figsize=(12,6))
cross_table = pd.crosstab(df["岗位大类"], df["标准化学历"])
cross_table.plot(kind="bar", stacked=True, color=color_list[:len(cross_table.columns)], ax=plt.gca())
plt.title("各岗位大类对应的学历需求分布", fontsize=13, pad=12)
plt.xlabel("岗位大类")
plt.ylabel("岗位数量")
plt.legend(title="学历门槛", bbox_to_anchor=(1.02, 1), loc="upper left", fontsize=8)
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
# 折线图 - 工作年限与平均薪资变化 
plt.figure(figsize=(9,4.5))
exp_sal = df.groupby("标准化工作年限")["平均月薪"].mean().sort_values()
plt.plot(exp_sal.index, exp_sal.values, marker="o", linewidth=2.2, color=color_list[1])
for x,y in zip(exp_sal.index, exp_sal.values):
    plt.text(x, y+300, f"{int(y)}", ha="center", fontsize=8)
plt.title("工作经验要求与岗位平均薪资趋势", fontsize=13, pad=12)
plt.xlabel("要求工作年限")
plt.ylabel("平均月薪（元）")
plt.grid(alpha=0.25)
plt.tight_layout()
plt.show()
# 热力图 - 城市梯队 × 薪资分层 交叉热力 
import seaborn as sns
plt.figure(figsize=(9,5))
heat_data = pd.crosstab(df["城市梯队"], df["薪资分层"])
sns.heatmap(heat_data, annot=True, fmt="d", cmap="Blues", alpha=0.85)
plt.title("城市梯队与薪资分层岗位数量热力图", fontsize=13, pad=12)
plt.tight_layout()
plt.show()
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.rcParams["figure.dpi"] = 120
plt.rcParams["axes.spines.top"] = False
plt.rcParams["axes.spines.right"] = False
color_list = ["#8DA9C4", "#D4A29C", "#B4C4B0", "#E8CFAF", "#A49CB7", "#D0A8B2"]
df = pd.read_csv("清洗后数据集.csv", encoding="utf-8-sig")
# 小提琴图 —— 不同城市梯队薪资分布 
plt.figure(figsize=(10,5))
sns.violinplot(x="城市梯队", y="平均月薪", data=df, palette=color_list[:3], inner="quartile")
plt.title("三大城市梯队薪资分布小提琴图", fontsize=13, pad=12)
plt.xlabel("城市梯队")
plt.ylabel("平均月薪（元）")
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
# 横向条形图 —— 各省份招聘岗位数量TOP10 
plt.figure(figsize=(11,6))
prov_cnt = df["省份"].value_counts().head(10)
bars = plt.barh(prov_cnt.index, prov_cnt.values, color=color_list[:len(prov_cnt)])
for bar in bars:
    width = bar.get_width()
    plt.text(width + 8, bar.get_y()+bar.get_height()/2, f"{int(width)}", va="center", fontsize=9)
plt.title("招聘岗位数量TOP10省份分布", fontsize=13, pad=12)
plt.xlabel("岗位数量")
plt.ylabel("省份")
plt.grid(axis="x", alpha=0.25)
plt.tight_layout()
plt.show()
# 箱线图 —— 不同学历薪资对比
plt.figure(figsize=(10,5))
edu_order = ["不限学历","大专","本科","硕士及以上","其他学历"]
edu_data = [df[df["标准化学历"]==edu]["平均月薪"] for edu in edu_order]
box = plt.boxplot(edu_data, labels=edu_order, patch_artist=True)
for patch, c in zip(box["boxes"], color_list[:4]):
    patch.set_facecolor(c)
    patch.set_alpha(0.7)
plt.title("不同学历门槛薪资分布箱线图", fontsize=13, pad=12)
plt.xlabel("学历要求")
plt.ylabel("平均月薪（元）")
plt.grid(axis="y", alpha=0.25)
plt.tight_layout()
plt.show()
```
